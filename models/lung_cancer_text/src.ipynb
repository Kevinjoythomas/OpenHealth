```python
df=pd.read_csv('./dataset.csv')
df
```

```python
df.info()
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#Removing Duplicates
df=df.drop_duplicates()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['GENDER']=le.fit_transform(df['GENDER'])
df['LUNG_CANCER'] = le.fit_transform(df['LUNG_CANCER'])
```

```python
df
```

```python
#to find the occurance
df['LUNG_CANCER'].value_counts()
```

```python
#dropping unnecessary features
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS_OF_BREATH'])
df_new
```

```python
#feature engineering to create new features from previous colenear ones
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
withoutCV=[]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
lr_model=RandomForestClassifier(random_state=0)
lr_model.fit(X_train, y_train)
y_lr_pred= lr_model.predict(X_test)
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
withoutCV.append(accuracy_score(y_test, y_lr_pred))
```

```python
y_lr_pred= lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
withoutCV.append(accuracy_score(y_test, y_lr_pred))
```
